```python
# Proyecto Kaggle - Exploración simple con CSV local


from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

print(" Sube tu archivo train.csv (guardado desde Excel o Kaggle):")
files.upload()

df = pd.read_csv("train.csv")
print(" Archivo cargado correctamente:", df.shape)
print("\nPrimeras filas del dataset:")
print(df.head())

print("\n Información general:")
print(df.info())

print("\n Estadísticas básicas:")
print(df.describe(include='all'))

if 'RENDIMIENTO_GLOBAL' in df.columns:
    df['RENDIMIENTO_GLOBAL'].value_counts().plot(kind='bar', color='skyblue')
    plt.title("Distribución del Rendimiento Global")
    plt.xlabel("Nivel")
    plt.ylabel("Cantidad de estudiantes")
    plt.show()
else:
    print("No se encontró la columna 'RENDIMIENTO_GLOBAL' en el dataset.")

if 'ESTU_PRGM_DEPARTAMENTO' in df.columns:
    df['ESTU_PRGM_DEPARTAMENTO'].value_counts().head(5).plot(kind='bar', color='orange')
    plt.title("Top 5 Departamentos con más estudiantes")
    plt.xlabel("Departamento")
    plt.ylabel("Cantidad")
    plt.show()
else:
    print("No se encontró la columna 'ESTU_PRGM_DEPARTAMENTO' en el dataset.")

print("\n Exploración terminada correctamente.")
```
